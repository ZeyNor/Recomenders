```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df_movies = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
df_credits = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
```

```python
tfidf =TfidfVectorizer(stop_words="english")
df_movies['overview'] = df_movies['overview'].fillna("")
tfidf_matrix = tfidf.fit_transform(df_movies['overview'])
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix )
```

```python
indices = pd.Series(df_movies.index, index = df_movies['original_title']).drop_duplicates()
indices
```

```python
indices['Avatar']
```

```python
def find(title,cosine_sim):
    idx = indices[title]
    sim_scores = enumerate(cosine_sim[idx])
    sim_scores = sorted(sim_scores, key=lambda x:x[1],reverse=True)
    sim_scores = sim_scores[1:11]

    sim_index = [i[0] for i in sim_scores]
    print(df_movies["original_title"].iloc[sim_index])

def get_recommendation(title, cosine_sim = cosine_sim):
    if not title in df_movies["original_title"].unique():
        full_title = []
        for (i,y) in enumerate(df_movies["original_title"].str.contains(title)):
            if y:
                full_title.append(df_movies["original_title"].iloc[i])
        print('We found following movies {}'.format(full_title))

        for t in full_title:
            find(t, cosine_sim)
            
    else:
        find(title, cosine_sim)

```

```python
get_recommendation('The Avengers')
```
